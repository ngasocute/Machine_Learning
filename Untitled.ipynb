```python

out = open("list_id.txt", "w")
listIDs = []
for file in [listFile[0]]:
  f = open(file, "r")
  print(f)
  urls = f.readlines()
  
  for url in urls:
    try:
      id = url.split('?v=')[1]
    except Exception as e:
      print(e)
    
    if id in listIDs:
      continue
    listIDs.append(id)
    print(id)
for id in listIDs:
  out.write(id)
print(len(listIDs))
```

```python
def getInfoYouTubeVideo(id):
  request = youtube.videos().list(
          part=youtubeVideoParts,
          id=id,
          regionCode="VN"
      )
  response = request.execute()
  return response
```
